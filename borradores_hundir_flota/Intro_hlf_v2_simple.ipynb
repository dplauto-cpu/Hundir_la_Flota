```python
import numpy as np
```

```python
#PASO 1, crear un tablero
def crear_tablero():
    """Crea un tablero de 10x10 lleno de '~'"""
    return np.full((10, 10), "~")
```

```python
#PASO 2, imprimir el tablero, si puede ser con coordenadas
def imprimir_tablero(tablero, ocultar_barcos=True):
    print("   " + " ".join([str(i) for i in range(10)])) # Xa mostrar núm. de columna
    print("   " + "-" * 20)
    
    for i, fila in enumerate(tablero):
        if ocultar_barcos:  # Ocultar barcos para el tablero de disparos
            fila_mostrar = ['~' if c == 'B' else c for c in fila]
        else:
            fila_mostrar = fila
        print(f"{i:2} | " + " ".join(fila_mostrar))

tablero_barcos = crear_tablero()
'''
enumerate porque permite permite recorrer iterable e indica índice y valor 
ocultar barcos es un booleano, False los muestra
for i >> i = índice de fila, fila = cada lista dentro de tablero
print núm. de fila, espacio, elementos unidos con espacio
'''
```

```python
#PASO 3, colocar barcos en tablero
def colocar_barcos(tablero):
    barcos = [(4, 1), (3, 2), (2, 3)] # barcos (tamaño, cantidad)
    posiciones_barcos = [
        [(0, 0, 0)],  # Horizontal fil 0, col 0-3
        [(2, 0, 0), (5, 7, 1)],  # Horizontal fil 2, col 0-2 / Vertical fil 5-7, col 7
        [(1, 8, 0), (4, 3, 0), (8, 5, 1)]  # Dos horizontales (fil 1 y 4), uno vertical (fil 8)
   
    ]
    # Posiciones predefinidas, ver si puedo hacerlas cambiar aleatoriamente x partida
    # Formato: [fila_inicio, col_inicio, orientación] (0=horizontal, 1=vertical)
    # Bucle xa colocar barcos por tamaño y cantidad
    for i, (tamaño, cantidad) in enumerate(barcos):
        for j in range(cantidad):
            pos = posiciones_barcos[i][j]
            fila, col, orientacion = pos
            
            for k in range(tamaño):
                if orientacion == 0:  # Horizontal
                    tablero[fila][col + k] = 'B'
                else:  
                    tablero[fila + k][col] = 'B'
    
    return tablero
```

```python
# PARA EJECUTAR SCRIPT
if __name__ == "__main__":
   
    tablero_barcos = crear_tablero()  # Crea el tablero
    
    tablero_barcos = colocar_barcos(tablero_barcos) # Coloca los barcos
    
    tablero_disparos = crear_tablero() # Crea tablero de disparos (sin barcos)
    
    print("=== TABLERO DE BARCOS ===") # tablero con barcos (para el jugador)
    imprimir_tablero(tablero_barcos, ocultar_barcos=False)
    
    print("\n=== TABLERO DE DISPAROS ===")
    imprimir_tablero(tablero_disparos, ocultar_barcos=True)
```
