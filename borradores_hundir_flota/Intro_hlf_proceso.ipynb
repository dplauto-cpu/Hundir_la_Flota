```python
 '''
 DEMO
Coloca un barco en el tablero representado por 'O' mayúsculas.
    barco (list): Lista de tuplas con las coordenadas del barco a colocar een tablero
    tablero (np.array): Tablero donde colocar el barco
0 <= fila < tablero.shape[0]: Fila debe ser mayor o igual a 0 y menor que el núm. total de filas.
0 <= columna < tablero.shape[1]: Columna debe ser mayor o igual a 0 y menor que el núm. total de columnas.
and: Ambas condiciones deben ser verdaderas para evitar posiciones fuera de rango.
    Returns:
        np.array: Tablero actualizado
    '''
fragata = [(1,1), (1,2)] 
portaviones = [(3,4), (3,5), (3,6), (3,7)]
tablero = np.full((10,10), " ")

# input que pida colocar fragata... en coordenadas (,)(,)(,)

def posicionar_barco(fragata, tablero):
    for fila, columna in fragata:
        if 0 <= fila < tablero.shape[0] and 0 <= columna < tablero.shape[1]:
            tablero[fila, columna] = "O"
    return tablero

tablero = posicionar_barco(fragata, tablero)

def posicionar_barco(portaviones, tablero):
    for fila, columna in portaviones:
        if 0 <= fila < tablero.shape[0] and 0 <= columna < tablero.shape[1]:
            tablero[fila, columna] = "O"
    return tablero


tablero = posicionar_barco(portaviones, tablero)
tablero_jugador_barcos = tablero
print(tablero_jugador_barcos)
```

```python
import random

def crear_tablero():
    """Crea un tablero vacío de 10x10"""
    return [['~' for _ in range(10)] for _ in range(10)]

def imprimir_tablero(tablero, ocultar_barcos=True):
    """Imprime el tablero con coordenadas"""
    print("   " + " ".join(str(i) for i in range(10)))
    for i, fila in enumerate(tablero):
        if ocultar_barcos:
            # Ocultar barcos para el tablero de disparos
            fila_mostrar = ['~' if c == 'B' else c for c in fila]
        else:
            fila_mostrar = fila
        print(f"{i}  " + " ".join(fila_mostrar))

def colocar_barco(tablero, tamano):
    """Coloca un barco aleatoriamente en el tablero"""
    while True:
        orientacion = random.choice(['H', 'V'])
        if orientacion == 'H':
            fila = random.randint(0, 9)
            col = random.randint(0, 9 - tamano)
            # Verificar espacio disponible
            if all(tablero[fila][col + i] == '~' for i in range(tamano)):
                for i in range(tamano):
                    tablero[fila][col + i] = 'B'
                return
        else:  # Vertical
            fila = random.randint(0, 9 - tamano)
            col = random.randint(0, 9)
            if all(tablero[fila + i][col] == '~' for i in range(tamano)):
                for i in range(tamano):
                    tablero[fila + i][col] = 'B'
                return

def colocar_barcos():
    """Coloca todos los barcos en el tablero"""
    tablero = crear_tablero()
    barcos = [4, 3, 3, 2, 2, 2]  # 1 portaaviones (4), 2 submarinos (3), 3 fragatas (2)
    for tamano in barcos:
        colocar_barco(tablero, tamano)
    return tablero

def disparar(tablero_barcos, tablero_disparos, fila, col):
    """Realiza un disparo y devuelve el resultado"""
    if tablero_disparos[fila][col] != '~':
        return "Ya disparaste aquí"
    
    if tablero_barcos[fila][col] == 'B':
        tablero_disparos[fila][col] = 'X'
        tablero_barcos[fila][col] = 'H'  # Marcar como hundido
        return "¡Tocado!"
    else:
        tablero_disparos[fila][col] = 'O'
        return "Agua"

def juego_terminado(tablero_barcos):
    """Verifica si todos los barcos han sido hundidos"""
    for fila in tablero_barcos:
        if 'B' in fila:
            return False
    return True

def main():
    print("=" * 40)
    print("     ¡BIENVENIDO A HUNDIR LA FLOTA!")
    print("=" * 40)
    
    # Colocar barcos aleatoriamente
    tablero_barcos = colocar_barcos()
    tablero_disparos = crear_tablero()
    
    # Contador de disparos
    disparos_realizados = 0
    
    while True:
        print("\n" + "=" * 40)
        print("TABLERO DE DISPAROS:")
        imprimir_tablero(tablero_disparos, ocultar_barcos=True)
        print("\n" + "=" * 40)
        
        try:
            fila = int(input("Ingresa la fila (0-9): "))
            col = int(input("Ingresa la columna (0-9): "))
            
            if fila < 0 or fila > 9 or col < 0 or col > 9:
                print("Coordenadas inválidas. Deben estar entre 0 y 9.")
                continue
            
            disparos_realizados += 1
            resultado = disparar(tablero_barcos, tablero_disparos, fila, col)
            print(f"\n{resultado}")
            
            if juego_terminado(tablero_barcos):
                print("\n" + "=" * 40)
                print("¡FELICIDADES! ¡HAS HUNDIDO TODOS LOS BARCOS!")
                print(f"Te tomó {disparos_realizados} disparos.")
                print("=" * 40)
                
                # Mostrar tablero final
                print("\nTABLERO FINAL:")
                imprimir_tablero(tablero_disparos, ocultar_barcos=False)
                break
                
        except ValueError:
            print("Por favor, ingresa números válidos.")

if __name__ == "__main__":
    main()
```

```python
Cómo funciona:
Tablero: Se crea un tablero de 10x10 con ~ (agua)

Barcos:

1 portaaviones (4 casillas)

2 submarinos (3 casillas cada uno)

3 fragatas (2 casillas cada una)

Colocación aleatoria: Los barcos se colocan automáticamente al inicio en posiciones aleatorias (horizontal o verticalmente)

Juego:

Muestra el tablero de disparos (oculta los barcos)

Pide coordenadas (fila y columna del 0 al 9)

Indica si tocaste un barco (X) o agua (O)

El juego termina cuando todos los barcos están hundidos

Características:

No puedes disparar dos veces en el mismo lugar

Muestra el número total de disparos al final

Interfaz simple por consola

Para jugar, simplemente ejecuta el script y sigue las instrucciones ingresando números del 0 al 9 para las coordenadas
```
