```python
#PASO 0,  
import numpy as np
import random
```

```python
#PASO 1, crear el tablero
def crear_tablero(): # copiar y pegar del visto en clase
        return np.full((10, 10), "~", dtype=str) # x si acaso todo str
```

```python
#PASO 2, imprimir el tablero
def imprimir_tablero(tablero, ocultar_barcos=True):
    print("   " + " ".join(str(i) for i in range(10)))  # Numeros de columnas
    for i in range(10):
        if ocultar_barcos:# Xa el tablero de disparos, a ver cómo implemento esto luego
            fila_mostrar = ['~' if tablero[i][j] == 'B' else tablero[i][j] for j in range(10)]
        else:
            fila_mostrar = [tablero[i][j] for j in range(10)]
        print(f"{i}  " + " ".join(fila_mostrar))  # Numero de fila y contenido


'''
    ocultar barcos es booleano, False muestra posición real, True el jugador ve agua
    for i >> i = índice de fila, fila = cada lista dentro de tablero
    print núm. de fila, espacio, elementos unidos con espacio
    
    Todo esto viene de un bucle clásico que sí entiendo:
    def imprimir_tablero(tablero, ocultar_barcos=True):
        print("   " + " ".join(str(i) for i in range(10)))
        for i in range(10):
            fila_mostrar = []
        for j in range(10):
            if ocultar_barcos and tablero[i][j] == 'B':
                fila_mostrar.append('~')  # Oculta barco, mola ~
            else:
                fila_mostrar.append(tablero[i][j])  # Muestra con
        print(f"{i}  " + " ".join(fila_mostrar)
'''
```

```python
#PASO 3, colocar barcos, intentar con uno y aleatorio, que el jugador no haga nada
def colocar_barco(tablero, tamano):
    while True:# en teoría ejecuta hasta encontrar posición válida ¿cómo lo paro?
        orientacion = random.choice(['H', 'V'])  # H: Horizontal, V: Vertical
        if orientacion == 'H': # H, la fila fija y la columna variable
            fila = random.randint(0, 9)
            col = random.randint(0, 9 - tamano)  # xa que quepa
            if all(tablero[fila][col + i] == '~' for i in range(tamano)):#sólo en casilla vacía
                for i in range(tamano):
                    tablero[fila][col + i] = 'B'  
                return
        else:  # V, columna fija y fila variable
            fila = random.randint(0, 9 - tamano)
            col = random.randint(0, 9)
            if all(tablero[fila + i][col] == '~' for i in range(tamano)):
                for i in range(tamano):
                    tablero[fila + i][col] = 'B'
                return
'''
    Coloca un barco de tamano especifico, no superpone.
    Crear función que coloque 6
'''
```

```python
#PASO 4, colocar todos los barcos del jugador
def colocar_barcos_jugador():
    tablero = crear_tablero()
    barcos = [4, 3, 3, 2, 2, 2]  
    for tamano in barcos:
        colocar_barco(tablero, tamano)
    return tablero
'''
    defino el tamaño de los 6 barcos a colocar, 1 de 4, 2 de 3, 3 de 2 [4, 3, 3, 2, 2, 2]
    Coloca los 6 barcos en tablero del jugador aleatoriamente.
 '''
```

```python
#PASO 5, coloca los barcos del contrincante
def colocar_barcos_ordenador():
    tablero = crear_tablero()
    barcos = [4, 3, 3, 2, 2, 2]
    for tamano in barcos:
        colocar_barco(tablero, tamano)
    return tablero
'''
   Coloca los 6 barcos en tablero del ordenador de aleatoriamente.
   Es igual que colocar_barcos_jugador
'''
```

```python
#PASO 6, disparos del jugador,
def disparar(tablero_barcos, tablero_disparos, fila, col):
    if tablero_disparos[fila][col] != '~': # xa no disparar dos veces en el mismo sitio
        return "ya disparado"
    if tablero_barcos[fila][col] == 'B': # # si hay barco en el tablero oculto
        tablero_disparos[fila][col] = 'X'  # tocado, marca X en tablero disparos
        tablero_barcos[fila][col] = 'H'    # xa que el jugador no se confunda 
        return "tocado"
    else:
        tablero_disparos[fila][col] = 'o'  # xa diferenciar el agua donde ya se ha disparado
        return "agua"

'''
    Coordenadas del disparo [fila] [columna], las introduce el jugador
    Debería ver algo así 
Antes
    Tablero barcos: ['B', '~'] o ['~', '~']
    Tablero disparos: ['~', '~']
Después de disparar
    Tablero barcos: [['H', '~'], ['~', '~']]
    Tablero disparos: [['X', '~'] o ['o', '~']]
Returns:
    String con el resultado del disparo
    """'''
```

```python
#PASO 7, los pepinos los dispara el ordenador, yo sí que flipo como un pepino
def disparo_ordenador(tablero_barcos_jugador, tablero_disparos_ordenador):
    while True:
        fila = random.randint(0, 9)
        col = random.randint(0, 9)
        if tablero_disparos_ordenador[fila][col] == '~': # ~ = nunca disparó aquí
            resultado = disparar(tablero_barcos_jugador, tablero_disparos_ordenador, fila, col)
            return fila, col, resultado
'''
    El ordenador elige coordenadas aleatorias,
    también usa la función disparar.
    if == "~" significa que nunca disparó aquí, me encanta su simplicidad.
'''
```

```python
#PASO 8, fin del juego, despedida y fin de fiesta
def juego_terminado(tablero_barcos):
    for i in range(10):
        for j in range(10):
            if tablero_barcos[i][j] == 'B':
                return False  # Todavia quedan barcos
    return True  # Todos los barcos hundidos
'''
 La batalla termina cuando todos los barcos han sido hundidos, no quedan 'B's en el tablero.
'''
```

```python
import random
import numpy as np
from utils import (
    crear_tablero,
    imprimir_tablero,
    colocar_barco,
    colocar_barcos_jugador,
    colocar_barcos_ordenador,
    disparar,
    juego_terminado,
    disparo_ordenador)

def main():
    print("=" * 60)
    print("     BIENVENIDO A HUNDIR LA FLOTA - MODO DOS JUGADORES")
    print("=" * 60)
    print("\nREGLAS:")
    print("- Tablero de 10x10 (coordenadas del 0 al 9), primera coordenada fila, segunda columna")
    print("- Tienes 6 barcos: 1 de tamaño 4, 2 de tamaño 3, 3 de tamaño 2")
    print("- El ordenador tiene los mismos barcos que tú")
    print("- Turnos alternos: empiezas disparando tú")
    print("- La batalla termina cuando un jugador hunde todos los barcos enemigos")
    input("\nPresiona ENTER para comenzar...")
    '''
    Funcion principal que controla el flujo del juego.
    Alterna turnos entre jugador y ordenador hasta que alguien gana.
    '''
    # Imprime los tableros
    print("\n" + "=" * 60)
    print("POR FAVOR ESPERE, COLOCANDO BARCOS...")
    print("=" * 60)
    
    # Tableros del jugador
    tablero_barcos_jugador = colocar_barcos_jugador()
    tablero_disparos_jugador = crear_tablero()  # Donde el jugador registra sus disparos al ordenador
    
    # Tableros del ordenador
    tablero_barcos_ordenador = colocar_barcos_ordenador()
    tablero_disparos_ordenador = crear_tablero()  # Donde el ordenador registra sus disparos al jugador
    
    print("Barcos colocados aleatoriamente para ti y el ordenador")
    
    # Estadisticas
    disparos_jugador = 0
    disparos_ordenador = 0
    
    # Bucle principal del juego
    turno_jugador = True  # El jugador empieza primero
    
    while True:
        if turno_jugador:
            # ===== TURNO DEL JUGADOR =====
            print("\n" + "=" * 60)
            print("TU TURNO - TABLERO DE DISPAROS (al ordenador):")
            print("=" * 60)
            imprimir_tablero(tablero_disparos_jugador, ocultar_barcos=True)
            
            print("\n" + "-" * 60)
            print("TUS BARCOS (posicion real):")
            print("-" * 60)
            imprimir_tablero(tablero_barcos_jugador, ocultar_barcos=False)
            
            # Solicitar coordenadas al jugador
            while True:
                try:
                    print("\n" + "-" * 60)
                    fila = int(input("Ingresa la fila (0-9): "))
                    col = int(input("Ingresa la columna (0-9): "))
                    
                    if fila < 0 or fila > 9 or col < 0 or col > 9:
                        print("Coordenadas invalidas. Deben estar entre 0 y 9.")
                        continue
                    
                    # Realizar disparo
                    resultado = disparar(tablero_barcos_ordenador, tablero_disparos_jugador, fila, col)
                    
                    if resultado == "ya_disparado":
                        print("Ya disparaste alli. Elige otra coordenada!")
                        continue
                    elif resultado == "tocado":
                        print("\nTOCADO!")
                    else:
                        print("\nAGUA")
                    
                    disparos_jugador += 1
                    break
                    
                except ValueError:
                    print("Por favor, ingresa numeros validos.")
            
            # Verificar si el jugador gano
            if juego_terminado(tablero_barcos_ordenador):
                print("\n" + "=" * 60)
                print("FELICIDADES! HAS HUNDIDO TODOS LOS BARCOS DEL ORDENADOR!")
                print(f"Te tomo {disparos_jugador} disparos.")
                print(f"El ordenador realizo {disparos_ordenador} disparos.")
                print("=" * 60)
                
                # Mostrar tableros finales
                print("\nTABLERO FINAL DEL ORDENADOR (tus disparos):")
                imprimir_tablero(tablero_disparos_jugador, ocultar_barcos=False)
                print("\nTABLERO FINAL DEL JUGADOR (disparos del ordenador):")
                imprimir_tablero(tablero_disparos_ordenador, ocultar_barcos=False)
                break
            
            # Cambiar turno
            turno_jugador = False
            input("\nFin de tu turno. Presiona ENTER para el turno del ordenador...")
            
        else:
            # ===== TURNO DEL ORDENADOR =====
            print("\n" + "=" * 60)
            print("TURNO DEL ORDENADOR")
            print("=" * 60)
            
            # Mostrar tablero del jugador con disparos recibidos
            print("\nTUS BARCOS (disparos recibidos del ordenador):")
            print("X = Tocado | O = Agua | B = Barco intacto | ~ = Sin disparar")
            print("-" * 60)
            imprimir_tablero(tablero_disparos_ordenador, ocultar_barcos=False)
            
            # El ordenador dispara
            fila, col, resultado = disparo_ordenador(tablero_barcos_jugador, tablero_disparos_ordenador)
            disparos_ordenador += 1
            
            print(f"\nEl ordenador dispara en ({fila}, {col})...")
            if resultado == "tocado":
                print("TE HA DADO EN UN BARCO!")
            else:
                print("El ordenador ha fallado.")
            
            # Verificar si el ordenador gano
            if juego_terminado(tablero_barcos_jugador):
                print("\n" + "=" * 60)
                print("GAME OVER! El ordenador ha hundido todos tus barcos.")
                print(f"Disparos del ordenador: {disparos_ordenador}")
                print(f"Tus disparos: {disparos_jugador}")
                print("=" * 60)
                
                # Mostrar tableros finales
                print("\nTABLERO FINAL DEL JUGADOR (tus barcos y disparos recibidos):")
                imprimir_tablero(tablero_disparos_ordenador, ocultar_barcos=False)
                break
            
            # Cambiar turno
            turno_jugador = True
            input("\nFin del turno del ordenador. Presiona ENTER para tu turno...")

if __name__ == "__main__":
    main()
```
